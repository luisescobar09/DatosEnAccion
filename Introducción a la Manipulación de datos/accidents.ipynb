# **Drop_duplicates**

# **Isnull**

```python
registro_nulos = {
  'Year': None,
  'Date Of Incident' : None,
  'Route' : 1,
  'Operator' : 'London General',
  'Group Name' : 'Go-Ahead',
  'Bus Garage' : 'Garage Not Available',
  'Borough' : 'Southwark',
  'Injury Result Description' : 'Injuries treated on scene',
  'Incident Event Type' : 'Onboard Injuries',
  'Victim Category' : None,
  'Victims Sex' : 'Male',
  'Victims Age' : 'Child'
}
df = df.append(registro_nulos, ignore_index = True)
```

```python
#Cambiar valores dentro de las columnas para practicar
df['Route'] = df['Route'].replace('(blank)', np.nan)

# Cambiar los valores 'Unknown' en 'Victims Age' a NaN
df['Victims Age'] = df['Victims Age'].replace('Unknown', np.nan)

# Cambiar los valores 'Unknown' en 'Victims Sex' a NaN
df['Victims Sex'] = df['Victims Sex'].replace('Unknown', np.nan)
df
```

```python
#Encontrar valores nulos
vacios = df.isnull().any()
vacios
```

# **Fillna**

```python
df['Route'].fillna('empty', inplace=True)
df
```

```python
df['Victims Sex'].fillna('Not specified', inplace=True)
df
```

```python
df['Victims Age'].fillna('Unknown', inplace=True)
df
```

# **Combinación de DataFrames**

```python
# Filtrar y ver solo la columna 'Victim Category'
Victim_Category = df['Victim Category']

print("Victim Category:")
print(Victim_Category)
```

```python
# Acceder a un grupo específico utilizando .loc()
Victim_Category = df.loc[df['Victim Category'] == 'Passenger', ['Victim Category']]
print(Victim_Category)
```

```python
#Conteo de victimas passenger
count_passenger = df['Victim Category'].value_counts()['Passenger']
print(count_passenger)
```

```python
#Victims Sex
Victims_Sex = df.loc[df['Victim Category'] == 'Passenger', ['Victim Category', 'Victims Sex']]
print(Victims_Sex)
```

```python
#Male
SexMale = df.loc[(df['Victims Sex'] == 'Male'), ['Victims Sex']]
print(SexMale)
```

```python
#Famale
SexFemale = df.loc[(df['Victims Sex'] == 'Female'), ['Victims Sex']]
print(SexFemale)
```

# Concat

```python
Victims_Sex = pd.concat([Victim_Category,SexFemale,SexMale],axis=1)
Victims_Sex = Victims_Sex.rename(columns={'Victims Sex': 'Sex'})
Victims_Sex_filled = Victims_Sex.fillna('Unknown')
print(Victims_Sex_filled)
```

# Merge

```python
# Combinar DataFrames Merge outer
merged_df = pd.merge(SexMale, SexFemale, how='outer')

print("DataFrame combinado:")
print(merged_df)
```

```python
#left
# Combinar DataFrames
merged_df = pd.merge(SexMale, SexFemale, how='left')

print("DataFrame combinado:")
print(merged_df)
```

```python
# Combinar DataFrames
merged_df = pd.merge(SexMale, SexFemale, how='right')

print("DataFrame combinado:")
print(merged_df)
```

# Join

```python
join_df = SexFemale.join(SexMale, lsuffix='_female', rsuffix='_male')
print(join_df)
```

```python
join_df = SexMale.join(SexFemale, lsuffix='_male', rsuffix='_female')
print(join_df)
```

```python
# Definir una función para calcular el conteo de casos con 'Female' en la columna 'Victims Sex'
def contar_female(df):
    conteo_female = (df['Victims Sex'] == 'Female').sum()
    conteo_df = pd.DataFrame({'Conteo de casos con Victims Sex = \'Female\'': [conteo_female]})
    return conteo_df
```

```python
# Definir una función para calcular el conteo de casos con 'Female' en la columna 'Victims Sex'
def contar_male(df):
    conteo_male = (df['Victims Sex'] == 'Male').sum()
    conteo_df = pd.DataFrame({'Conteo de casos con Victims Sex = \'Male\'': [conteo_male]})
    return conteo_df
```

```python
resultado = contar_male(df)
print(resultado)
```

```python
resultado = contar_female(df)
print(resultado)
```
